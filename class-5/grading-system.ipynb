```python
from typing import List,Dict

def get_valid_marks(subject_name: str) -> float:
    while True:
        try:
            marks = float(input(f"Enter marks for {subject_name}: "))
            if 0 <= marks <= 100:
                return marks
            else:
                print("Marks should be between 0 and 100")
        except ValueError:
            print("Please enter a valid number")

def calculate_grade(percentage: float) -> str:
    if percentage >= 90:
        return 'A+'
    elif percentage >= 80:
        return 'A'
    elif percentage >= 70:
        return 'B+'
    elif percentage >= 60:
        return 'B'
    elif percentage >= 50:
        return 'C'
    elif percentage >= 40:
        return 'D'
    else:
        return 'F'

def main():
    # Get student information
    name = input("Enter student's name: ").upper()
    roll_no = input("Enter roll number: ")
    
    # Subject list
    subjects: List[str] = ['ENGLISH', 'PHYSICS', 'CHEMISTRY', 'MATHEMATICS', 'COMPUTER SCIENCE']
    subject_marks: Dict[str, float] = {}
    total_obtained: float = 0
    
    
    # Get marks for each subject
    print("\nEnter marks for each subject (out of 100):")
    for subject in subjects:
        marks = get_valid_marks(subject)
        subject_marks[subject] = marks
        total_obtained += marks
    
    # Calculate percentage and overall grade
    total_marks: int = len(subjects) * 100
    percentage: float = (total_obtained / total_marks) * 100
    overall_grade: str = calculate_grade(percentage)
    
    # Print Result Card
    print("\n" + "="*60)
    print(f"{' '*20}MARKS CERTIFICATE")
    print("="*60)
    
    # Header Information
    print(f"Name: {name}")
    print(f"Roll Number: {roll_no}")
    print("\n" + "-"*60)
    
    # Subject wise marks and grades
    print(f"{'SUBJECT':<20} {'MAX MARKS':<15} {'OBTAINED':<15} {'GRADE':<10}")
    print("-"*60)
    
    for subject, marks in subject_marks.items():
        subject_grade = calculate_grade(marks)
        print(f"{subject:<20} {100:<15} {marks:<15} {subject_grade:<10}")
    
    print("-"*60)
    
    # Overall Result
    print(f"\nTotal Marks: {total_marks}")
    print(f"Marks Obtained: {total_obtained}")
    print(f"Percentage: {percentage:.2f}%")
    print(f"Overall Grade: {overall_grade}")
    
    # Result Remarks
    status:str = "PASS" if percentage >= 40 else "FAIL"
    print(f"Remarks: {status}")
    print("="*60)
    
    # Developer Information
    print("\n" + "*"*60)
    print(f"{' '*15}Developed by Muhammad Aasim ")
    print(f"{' '*15}PIAIC112158 Islamabad Batch 68")
    print("*"*60)

main
```
